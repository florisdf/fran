```python
from datasets.fran_dataset import FRANDataset
```

```python
ds_train = FRANDataset(data_root='../data/FRAN_dataset/', is_val=False)
```

```python
import seaborn as sns
import numpy as np

df = ds_train.df

med_ages = df.groupby('label').agg('median')['age']

df['centered_age'] = df.apply(lambda row: row['age'] - med_ages[row['label']], axis=1)

sns.displot(df, x='centered_age')
```
